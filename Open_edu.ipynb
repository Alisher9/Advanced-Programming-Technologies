```python
import requests
import re # regular expressions
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
CHROME_DRIVER_PATH = r'C:\Users\FLOX\ChromeDriver\chromedriver.exe'
driver = webdriver.Chrome(CHROME_DRIVER_PATH)
```

```python
# step 1
URL = 'https://www.open.edu/openlearn/free-courses/full-catalogue'
```

```python
driver.get(URL)
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
```

```python
names = []
time = []
for block in soup.findAll('table', attrs={'class':'views-table cols-3 sortable'}):
    name_str = block.find('tr', attrs={'class': 'odd'}).text
    time_str = block.find('tr', attrs={'class': 'even'}).text
    names.append(name_str)
    time.append(time_str)
```

```python
data = {
    'Name1': names,
    'Name2': time
}
df = pd.DataFrame(data)
```

```python
df = df.replace('\n\n','- ', regex=True)
```

```python
data=df.join(df['Name1'].str.split('-', 1, expand=True).rename(columns={0:'Name', 1:'Description'}))
```
