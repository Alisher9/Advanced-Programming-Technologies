```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

### Longest Courses on skillbox


```python
df_stepik = pd.read_excel('All_courses.xlsx',sheet_name=3)
longest_stepik = df_stepik.sort_values('Count_of_Month/Duration',ascending = False)
longest_stepik_top = longest_stepik.head(15)

figure(figsize=(6,4), dpi=80)

plt.xlabel('Course Name')
plt.ylabel('Course length in Months')

plt.xticks(rotation='vertical')

bars = plt.bar(longest_stepik_top['Name'], longest_stepik_top['Count_of_Month/Duration'])

plt.savefig('longest_courses_skillbox.png')
```

### Average Course Length in Skillbox


```python
avg_skillbox = df_stepik['Count_of_Month/Duration'].mean()
rounded_avg_skillbox = round(avg_skillbox)

print('Average course length in Skillbox is : {}'.format(rounded_avg_skillbox))
```
